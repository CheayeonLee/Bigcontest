# Y변수 EDA

```python
# 데이터 불러오기
import pandas as pd
data=pd.read_excel('AllMerge.xlsx')
data
```

```python
import matplotlib.pyplot as plt

plt.hist(data['MCT_ME_D'].dropna(), bins=20, edgecolor='black')
plt.xlabel('MCT_ME_D')
plt.ylabel('Frequency')
plt.title('Histogram of MCT_ME_D')
plt.show()
```

```python
data[['MCT_ME_D','RC_M1_SAA','RC_M1_UE_CUS_CN']]
```

```python
closed=data[data['MCT_ME_D'].notnull()]
```

```python
closed[['RC_M1_SAA','RC_M1_UE_CUS_CN']].describe()
```

## 폐업한 가맹점의 전체 월별 데이터 분포

```python
import matplotlib.pyplot as plt

plt.hist([closed['RC_M1_SAA'],
          closed['RC_M1_UE_CUS_CN']], bins=20, edgecolor='black')
plt.xlabel('sales & customer(closed)')
plt.ylabel('Frequency')
plt.title('Histogram of MCT_ME_D')
plt.show()
```

## 폐업한 가맹점 중 가장 최근 월의 데이터

```python
# 폐업한 가맹점 중에 가장 최근 데이터만 뽑아서 확인
latest_closed = closed.loc[closed.groupby('ENCODED_MCT')['TA_YM'].idxmax()]
latest_closed
```

```python
import matplotlib.pyplot as plt

# TA_YM별 평균 계산
trend = latest_closed.groupby('TA_YM')[['RC_M1_SAA','RC_M1_UE_CUS_CN']].mean()

# 선 그래프로 시각화
trend.plot(marker='o')
plt.xlabel('TA_YM')
plt.ylabel('Average Value')
plt.title('Average Sales & Customers by TA_YM')
plt.show()
```

```python
for ta, group in latest_closed.groupby('TA_YM'):
    plt.figure()
    plt.hist([group['RC_M1_SAA'],
              group['RC_M1_UE_CUS_CN']],
             bins=20, edgecolor='black', label=['Sales','Customers'])
    plt.title(f'Histogram for {ta}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
latest_closed['MCT_ME_D'] = latest_closed['MCT_ME_D'].astype(str).str.split('.').str[0]
latest_closed['closed_YM'] = latest_closed['MCT_ME_D'].str[:6]
latest_closed['closed_YM']
```

# Merge

```python
import pandas as pd
data=pd.read_excel('AllMerge_final(7개 카테고리 추가).xlsx')
data=data.drop('Unnamed: 0',axis=1)
data
```

```python
data
```

## 외부 데이터들 불러와서 합치기

```python
eco=pd.read_excel('경제적지표.xlsx')
eco
```

```python
merged_df1 = pd.merge(data, eco, left_on='TA_YM', right_on='전국', how='left')
merged_df1
```

```python
product=pd.read_excel('생산자물가지수.xlsx')
product
```

```python
product=product.drop('구분',axis=1)
product
```

```python
product_melt = pd.melt(product, id_vars=['업종'], var_name='연도', value_name='product_price')
```

```python
merged_df2 = pd.merge(merged_df1,  product_melt,  left_on=['TA_YM', 'HPSN_MCT_ZCD_NM'], right_on=['연도', '업종'], how='left')
merged_df2
```
